```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Define the labels for the classes
labels = ['Atelectasis', 'Effusion', 'Infiltration', 'No Finding', 'Nodule', 'Pneumothorax']

# Define the values of the confusion matrix
conf_matrix = np.array([[100, 10, 5, 0, 0, 0],
                        [10, 90, 5, 5, 0, 0],
                        [5, 5, 80, 10, 5, 0],
                        [0, 5, 10, 85, 0, 0],
                        [0, 0, 5, 0, 90, 5],
                        [0, 0, 0, 0, 5, 95]])

# Calculate the accuracy of each class
accuracy = np.diag(conf_matrix) / np.sum(conf_matrix, axis=1)

# Calculate the misdiagnosis of each class
misdiagnosis = 1 - accuracy

# Create a heatmap of the confusion matrix
fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(conf_matrix, cmap='coolwarm')
# Add the class labels to the x and y axes
ax.set_xticks(np.arange(len(labels)))
ax.set_yticks(np.arange(len(labels)))
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)


# Add the values of the confusion matrix to the heatmap
for i in range(len(labels)):
    for j in range(len(labels)):
        text = ax.text(j, i, conf_matrix[i, j],
                       ha='center', va='center', color='w')

# Add a colorbar to the heatmap
cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.set_ylabel('Number of Samples', rotation=-90, va='bottom')

# Add a title to the heatmap
ax.set_title('Confusion Matrix')

# Show the heatmap
plt.show()

# Create a bar chart of the accuracy of each class
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(labels, accuracy, color='navy')

# Add the percentage values of the accuracy to the bars
for i, v in enumerate(accuracy):
    ax.text(i, v, f'{v:.2%}', ha='center', va='bottom')

# Add a title and labels to the bar chart
ax.set_title('Accuracy of Each Class')
ax.set_xlabel('Class')
ax.set_ylabel('Accuracy')

# Show the bar chart
plt.show()

# Create a bar chart of the misdiagnosis of each class
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(labels, misdiagnosis, color='navy')

# Add the percentage values of the misdiagnosis to the bars
for i, v in enumerate(misdiagnosis):
    ax.text(i, v, f'{v:.2%}', ha='center', va='bottom')

# Add a title and labels to the bar chart
ax.set_title('Misdiagnosis of Each Class')
ax.set_xlabel('Class')
ax.set_ylabel('Misdiagnosis')

# Show the bar chart
plt.show()
```
